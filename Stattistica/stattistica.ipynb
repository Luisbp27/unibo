```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import csv file
df = pd.read_csv('data_thigh.csv')

# Configurate data frame to show all columns
pd.set_option('display.max_columns', None)

df.head()
```

```python
corr_fig = plt.figure(figsize=(20, 18))

# Plot correlation matrix
sns.heatmap(df.corr(), annot=True, square=True, cmap='coolwarm', linewidths=0.5)
```

```python
# Delete some columns
df.drop([
    'Severe_Thigh_Injuries (1=yes, 0=no)',
    'Noncontact_Thigh_Injuries(1=yes, 0=no)',
    'Knowledgeability (1=yes, 2=no)',
    'Prevention Measure Stretching (1=yes, 0=no)',
    'Prevention Measure Warm Up (1=yes, 0=no)',
    'Prevention Measure Specific Strength Exercises (1=yes, 0=no)',
    'Prevention Measure Bracing (1=yes, 0=no)',
    'Prevention Measure Taping (1=yes, 0=no)',
    'Prevention Measure Shoe Insoles (1=yes, 0=no)',
    'Prevention Measure Medical Corset (1=yes, 0=no)'
    ], axis=1, inplace=True)

df.head()
```

```python
corr_fig = plt.figure(figsize=(10, 8))

# Plot correlation matrix
sns.heatmap(df.corr(), annot=True, square=True, cmap='coolwarm', linewidths=0.5)
```

```python
# Calculate the probability of a thigh injury
df['Number of Thigh Injuries'] = df['Number of Thigh Injuries'].replace(2, 0)
df['Number of Thigh Injuries'].value_counts()

# Plot the probability of a thigh injury
ax = sns.countplot(x='Number of Thigh Injuries', data=df)
ax.set_title('Probability of a Thigh Injury')
ax.set_ylabel('Probability')
ax.set_ylim([0, 100])
plt.show()
```

```python
# Count the number of thigh injuries
thigh_injuries_count = df['Number of Thigh Injuries'].value_counts()

# Divide the count of thigh injuries by the total number of injuries
mean = thigh_injuries_count / df['Number of Injuries'].count() * 100

mean
```

```python
# Calcula la mediana de las lesiones de muslo
median = df['Number of Thigh Injuries'].median()
median
```

```python
# Calcula la moda de las lesiones de muslo
mode = df['Number of Thigh Injuries'].mode()
mode
```

```python
# Calcular las medidas de dispersión de las lesiones de muslo
# Rango
range_ = df['Number of Thigh Injuries'].max() - df['Number of Thigh Injuries'].min()
range_
```

```python
# Varianza
variance = df['Number of Thigh Injuries'].var()
variance
```

```python
# Desviación estándar
std = df['Number of Thigh Injuries'].std()
std
```

```python
# Histograma de las lesiones de muslo por intervalos frente a las lesiones previas
ax = sns.histplot(data=df, x='Number of Thigh Injuries', bins=3)
ax.set_title('Thigh Injuries Histogram')
ax.set_ylabel('Number of players')
plt.show()
```

```python
# Coeficiente de pearson
pearson = df['Number of Thigh Injuries'].corr(df['Number of Injuries'])
pearson
```

```python
# Coeficiente de determinación de las lesiones de muslo
r_squared = pearson ** 2
r_squared
```

```python
# Agrupa las lesiones totales por edad
age_group = df.groupby('Age')['Number of Thigh Injuries'].sum()

# Grafiaca las lesiones totales por edad
ax = age_group.plot(kind='bar', figsize=(10, 8))
ax.set_title('Number of Thigh Injuries by Age')
ax.set_xlabel('Age')
ax.set_ylabel('Number of Thigh Injuries')
plt.show()
```

```python
# Extraer los valores de age_group a una lista
lesiones_muslo = age_group.values.tolist()
lesiones_muslo

# Crea una lista de edades entre 13 y 35 años
edades = list(range(13, 36))

# Coeficiente de pearson entre lesiones_muslo y la edades
pearson = pd.Series(lesiones_muslo).corr(pd.Series(edades))
pearson
```

```python
# Covarianza entre lesiones_muslo y la edades
covariance = pd.Series(lesiones_muslo).cov(pd.Series(edades))
covariance
```

```python
# Coeficiente de determinación entre lesiones_muslo y la edades
r_squared = pearson ** 2
r_squared
```

```python
# Modelo lineal entre lesiones_muslo y la edades (listas)
# Ajuste de un modelo lineal
coefficients = np.polyfit(edades, lesiones_muslo, 1)
poly = np.poly1d(coefficients)

# Crear un conjunto de puntos para la línea de regresión
x_line = np.linspace(min(edades), max(edades), 100)
y_line = poly(x_line)

# Graficar los datos y la línea de regresión
plt.scatter(edades, lesiones_muslo)
plt.plot(x_line, y_line, color='red')
plt.xlabel('Ages')
plt.ylabel('Number of Thigh Injuries')
plt.show()
```

```python
# Modelo exponencial entre lesiones_muslo y la edades
# Ajuste de un modelo exponencial
coefficients = np.polyfit(edades, lesiones_muslo, 2)
poly = np.poly1d(coefficients)

# Crear un conjunto de puntos para la línea de regresión
x_line = np.linspace(min(edades), max(edades), 100)
y_line = poly(x_line)

# Graficar los datos y la línea de regresión
plt.scatter(edades, lesiones_muslo)
plt.plot(x_line, y_line, color='red')
plt.xlabel('Ages')
plt.ylabel('Number of Thigh Injuries')
plt.show()

poly
```

```python
# Realizar predicciones
edades_prediccion = [18, 20, 30]  # Puedes ajustar las edades que desees predecir
lesiones_prediccion = poly(edades_prediccion)

# Graficar los datos, la línea de regresión y las predicciones
plt.scatter(edades, lesiones_muslo, label='Real data')
plt.plot(x_line, y_line, color='red', label='Linear Regression')
plt.scatter(edades_prediccion, lesiones_prediccion, color='green', label='Predictions', marker='x')
plt.xlabel('Ages')
plt.ylabel('Number of Thigh Injuries')
plt.legend()
plt.show()

# Mostrar las predicciones
for edad, prediccion in zip(edades_prediccion, lesiones_prediccion):
    print(f"Para la edad {edad}, la predicción es {prediccion}")
```
